## Preprocessing

### lowercase, stopwords, tokenization, special characters, stemming/lemmi, 
### Polarity, word count, sentence length, etc
### Stemming — stemming is the process of reducing inflected words to their word stem, base, or root form
### Lemmatization — process of grouping together the inflected forms of a word so they can be analyzed as a single item, identified by the word’s lemma, or dictionary form.

```python
stops = set(stopwords.words("english"))
def cleanData(text, lowercase = False, remove_stops = False, stemming = False):
    txt = str(text)
    txt = re.sub(r'[^A-Za-z0-9\s# \n]',r'',txt)
#     txt = re.sub(r'\n',r' ',txt)
    
    if lowercase:
        txt = " ".join([w.lower() for w in txt.split()])
        
    if remove_stops:
        txt = " ".join([w for w in txt.split() if w not in stops])
    
    if stemming:
        st = PorterStemmer()
        txt = " ".join([st.stem(w) for w in txt.split()])
        
    return txt
```

```python
def feature_enginerring(train):
    
    train['cleaned'] = train['text'].map(lambda x: cleanData(x, lowercase=True, remove_stops=True, stemming=False))
    train['length'] = train['cleaned'].map(lambda x: len(x))
    train['word_count'] = train['cleaned'].map(lambda x: len(nltk.word_tokenize(x)))
    train['polarity'] = train['cleaned'].map(lambda x: TextBlob(x).sentiment.polarity)
    
    return train
```

### Note you are removing all special character while you can use some of them like #hashtags, https, URLs, @mentions, etc

```python
train = feature_enginerring(train)
train.head()
```

### Matplotlib vs Seaborn vs plotly

### Seaborn has sharp features but are limitng in capablities compared to matplotlib. If you want to create the same histogram, but with the two variables stacked next to each other as opposed to overlaid, you would need to fall back to matplotlib
### Plotly graphs are automatically outfitted with hover tool capabilities — hovering your mouse over any of the bars of data will display the numerical values. Plus you can download (sightly interactive)

```python
fig = make_subplots(rows=3, cols=1)
train['polarity'].iplot(
    kind='hist',
    bins=50,
    theme = 'solar',
    dimensions=(1000,500),
    xTitle='polarity',
    linecolor='white',
    yTitle='count',
    title='Sentiment Polarity Distribution')
train['word_count'].iplot(
    kind='hist',
    bins=50,
    dimensions=(1000,500),
    xTitle='polarity',
    linecolor='black',
    yTitle='count',
    title='Word Distribution')
train['length'].iplot(
    kind='hist',
    bins=100,
    dimensions=(1000,500),
    xTitle='polarity',
    linecolor='black',
    yTitle='count',
    title='Sentence length Distribution')
```

```python
# now bar plots for catgories
train.groupby('target').count()['id'].iplot(kind='bar',
    xTitle='Disaster',
    colors="dodgerblue",
    bargap=0.5, #gap b/w bars
    dimensions=(500,500),
    theme="solar", #background
    linecolor='white', #axes line color
    orientation ='h', #vertical or horizontal
    yTitle='count',
    title='Disaster vs Non-Disaster Distribution')
train.groupby('keyword').count()['id'].iplot(kind='bar',
    xTitle='Keywords',
    colors ='green',
    dimensions=(1000,500),
    theme ='ggplot',
    linecolor='black',
    yTitle='count',
    title='Keyword Distribution')
```
